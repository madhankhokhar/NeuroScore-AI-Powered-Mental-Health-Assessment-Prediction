```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='gender', y='bmi', data=data)
plt.title('Boxplot of BMI by Gender')
plt.xlabel('Gender')
plt.ylabel('BMI')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='depression_severity', hue='gender')
plt.title('Count of Depression Severity by Gender')
plt.xticks(rotation=45)
plt.xlabel('Depression Severity')
plt.ylabel('Count')
plt.legend(title='Gender')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='anxiety_severity', y='phq_score', data=data)
plt.title('PHQ Score by Anxiety Severity')
plt.xlabel('Anxiety Severity')
plt.ylabel('PHQ Score')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
data['depression_diagnosis'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Depression Diagnosis Distribution')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='depression_severity', hue='anxiety_treatment')
plt.title('Count of Anxiety Treatment by Depression Severity')
plt.xticks(rotation=45)
plt.xlabel('Depression Severity')
plt.ylabel('Count')
plt.legend(title='Anxiety Treatment')
plt.show()
```

```python
features = data.select_dtypes(include=['float64', 'int64']).columns.tolist()
```

```python
num_data = data[features]
```

```python
correlation_matrix = num_data.corr()
```

```python
print(correlation_matrix)
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .8})
```
